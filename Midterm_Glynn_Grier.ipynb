```python
import pandas as pd
df= pd.read_csv('amazon_data.csv')
df.shape
```

```python
df.head()
```

```python
df= df.fillna(0)
df.head()
```

```python
df["Total Charged"] = df["Total Charged"].str.replace('$','').astype(float)
df.head()
```

```python
df["Total Charged"].sum()
```

```python
df["Total Charged"].mean()
```

```python
df["Total Charged"].median()
```

```python
df["Total Charged"].mode()
```

```python
df["Total Charged"].max()
```

```python
df["Total Charged"].min()
```

```python
import numpy as np
stdev= sum(df["Total Charged"])/len(df["Total Charged"])
print(stdev)
```

```python
df["Tax Charged"] = df["Tax Charged"].str.replace('$','').astype(float)
df.head()
```

```python
df["Tax Charged"].sum()
```

```python
df["Tax Charged"].mean()
```

```python
df["Tax Charged"].mode()
```

```python
df["Tax Charged"].median()
```

```python
df["Tax Charged"].max()
```

```python
df["Tax Charged"].min()
```

```python
df.plot.bar(x='Tax Charged', y='Total Charged', figsize=(20,10))
```
