```python
import matplotlib.pyplot as plt
import numpy as np
marks ={'Andy':88, 'Amy':66, 'James':90, 'Jules':55, 'Arthur':77}
name=list(marks.keys())
grade= list(marks.values())
plt.title('Student Mark')
plt.bar(range(len(marks)), grade, tick_label=name)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
marks={'Andy':88, 'Amy':66, 'James':90, 'Jules':55, 'Arthur':77}
name=list(marks.keys())
grade= list(marks.values())
plt.pie(grade, labels=name)
plt.legend(title= 'Student Mark', bbox_to_anchor=(1.05, 1.0), loc= 'upper right')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x= np.array(['A', 'B', 'C', 'D'])
y= np.array([7, 4, 9, 13])
plt.subplot(2, 3, 1)
plt.title= ('Bar Plot')
plt.bar(x, y)

y=np.array([10, 20, 30, 40])
plt.subplot(2, 3, 2)
plt.title= ('Pie Chart')
plt.pie(y)


x= np.array([1, 2, 3, 4, 5])
y= np.array([10, 12, 14, 15, 17])
plt.subplot(2,3,3)
plt.title=('Grid plot')
plt.plot(x, y)
plt.grid()

x1= np.array([0, 4])
y1=([1, 5])
x2=([0, 4])
y2= ([5, 9])
x3= ([0, 4])
y3=([9, 13])
plt.subplot(2, 3, 4)
plt.title=('linear')
plt.plot(x1, y1, x2, y2, x3, y3,)

x= np.array([2, 3, 4, 6, 8, 11, 4])
y= np.array([4, 5, 3, 7, 1, 12, 5])

plt.subplot(2, 3, 5)
plt.title= ('Scatter plot')
plt.scatter(x, y)

x= np.random.normal(120, 20, 140)
plt.subplot(2, 3, 6)
plt.title=('Histogram')
plt.hist(x)

plt.show()
```
